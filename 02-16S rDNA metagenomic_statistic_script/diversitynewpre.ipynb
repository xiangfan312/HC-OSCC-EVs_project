```python
import pandas as pd
sample_md = pd.DataFrame([
    ['HC-1-pre','pre-culture'],
    ['HC-2-pre','pre-culture'],
    ['HC-3-pre','pre-culture'],
    ['OSCC-1-pre','pre-culture'],
    ['OSCC-2-pre','pre-culture'], 
    ['OSCC-3-pre','pre-culture'], 
    ['OSCC-4-pre','pre-culture'],
    ['HC-1-post','post-culture'],
    ['HC-2-post','post-culture'],
    ['HC-3-post','post-culture'],
    ['OSCC-1-post','post-culture'],
    ['OSCC-2-post','post-culture'],
    ['OSCC-3-post','post-culture'],
    ['OSCC-4-post','post-culture']],
    index=[ 'HC-1-pre', 'HC-2-pre','HC-3-pre','OSCC-1-pre','OSCC-2-pre','OSCC-3-pre','OSCC-4-pre','HC-1-post','HC-2-post',
           'HC-3-post','OSCC-1-post','OSCC-2-post', 'OSCC-3-post',  'OSCC-4-post'],
    columns=['SimpleId','Group'])
sample_md['Adiv_shannon'] = adiv_shannon
sample_md
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator
palette = ["lightskyblue",  "mistyrose"]
tips = sns.load_dataset("tips")
order=['pre-culture','post-culture']
# 绘制箱线图和点图
ax = sns.boxplot(x='Group', y='Adiv_shannon', data=sample_md, order=order, palette=palette)

xlabel = plt.xlabel('Group', fontsize=15)
plt.xticks(fontsize=9)
sns.stripplot(x='Group', y='Adiv_shannon', data=sample_md, size=2, jitter=True, edgecolor="black", linewidth=2, order=order)

# 添加注释
annot = Annotator(ax, [('pre-culture','post-culture')],
                  data=sample_md, x='Group', y='Adiv_shannon', order=order)
annot.configure(loc='outside')
annot.set_custom_annotations(["p=0.479011"])
annot.configure(test='t-test_ind', text_format='full',fontsize='large', loc='inside',
                comparisons_correction=None, line_height=0.02, text_offset=8)
annot.annotate()
ylabel = plt.ylabel('Shannon Diversity', fontsize=15)
# 调整横坐标字体大小
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#添加标题HC vs OSCC
plt.title('Shannon Diversity of Pre-culture vs Post-culture', fontsize=15)
#y轴坐标范围
plt.ylim(2, 4.8)
# 保存图像
plt.savefig('Shannonprepost.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
