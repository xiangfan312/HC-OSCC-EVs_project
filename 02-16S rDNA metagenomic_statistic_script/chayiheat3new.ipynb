```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import os
import PyComplexHeatmap as pchm
from PyComplexHeatmap import *
from matplotlib.colors import Normalize
```

```python
row_ha_left = HeatmapAnnotation(
                            Change =anno_simple(data_rows['Change'],colors={'up': 'red', 'down': 'green'},
                                                add_text=True,legend=False,height=5,
                                                text_kws={'color':'white', 'fontsize':12,'fontweight':'bold',
                                                          'rotation':-90,'horizontalalignment':'left'}),
                            label_kws={'fontsize':15,'color':'black','rotation':-45,'horizontalalignment':'left'},
                            label_side='bottom',axis=0,verbose=0
                            )

col_ha = HeatmapAnnotation(Group=data_cols.Group,label_side='right', axis=1,label_kws={'fontsize': 15})
plt.figure(figsize=(12, 12))
cm = ClusterMapPlotter(data=data0,left_annotation=row_ha_left,top_annotation=col_ha,
                       show_rownames=True, show_colnames=True,col_names_side='bottom',
                       row_split=data_rows.Change, cmap='YlGnBu',
                       rasterized=True, legend=True,
                       xticklabels_kws={'labelrotation':-20,'labelsize': 15}, yticklabels_kws={'labelsize': 15},row_dendrogram=True,col_dendrogram=True)
plt.title('OSCC-pre-culture vs OSCC-post-culture', fontsize=23, loc='center', x=-54, y=4)
plt.savefig('heatmap of OSCC-pre-culture vs OSCC-post-culture1.pdf',dpi=300,bbox_inches='tight')
plt.show()
```
