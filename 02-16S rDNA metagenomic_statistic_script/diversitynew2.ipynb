```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
#读取文件
df1 = pd.read_csv('newples2.txt',sep='\t')
df1
```

```python
#df1以groups为索引
df2 = df1.set_index('groups')
df2
```

```python
#df2索引相同的行合并
df3 = df2.groupby('groups').sum()
df3
```

```python
#提取df3的'HC-1-pre', 'HC-2-pre','HC-3-pre','OSCC-1-pre','OSCC-2-pre','OSCC-3-pre','OSCC-4-pre','HC-1-post','HC-2-post','HC-3-post','OSCC-1-post','OSCC-2-post', 'OSCC-3-post',  'OSCC-4-post'
df4 = df3[['HC-1-pre', 'HC-2-pre','HC-3-pre','OSCC-1-pre','OSCC-2-pre','OSCC-3-pre','OSCC-4-pre','HC-1-post','HC-2-post','HC-3-post','OSCC-1-post','OSCC-2-post', 'OSCC-3-post',  'OSCC-4-post']]
df4
```

```python
#提取df2的'HC-1-pre', 'HC-2-pre','HC-3-pre','OSCC-1-pre','OSCC-2-pre','OSCC-3-pre','OSCC-4-pre'
da= df4[['HC-1-pre', 'HC-2-pre','HC-3-pre','OSCC-1-pre','OSCC-2-pre','OSCC-3-pre','OSCC-4-pre']]
da
```

```python
#da5保存为pre1.txt
da.to_csv('pre1.txt',sep='\t')
```

```python
#df2提取'HC-1-post','HC-2-post','HC-3-post','OSCC-1-post','OSCC-2-post', 'OSCC-3-post',  'OSCC-4-post'
dg = df4[['HC-1-post','HC-2-post','HC-3-post','OSCC-1-post','OSCC-2-post', 'OSCC-3-post',  'OSCC-4-post']]
dg
```

```python
#dg4保存为post1.txt
dg.to_csv('post1.txt',sep='\t')
```

```python
df5=df4
```

```python
df6= pd.DataFrame(df5.values.T,columns = df5.index,index = df5.columns) 
df6.head()
```

```python
numpy_array=np.array(df6.values)
arr=numpy_array
arr

```

```python
#提取df5的索引
dh = df5.index
array = np.array(dh)  # 将df转换为list对象
otu_ids = array.tolist()  # 获取第一列的数据
print(otu_ids)
```

```python
#df6的列名
simple_ids= df5.columns
print(simple_ids)
```

```python
from skbio.diversity import alpha_diversity
from skbio.diversity.alpha import shannon
adiv_obs_otus = alpha_diversity('observed_otus', arr,simple_ids)
adiv_obs_otus
```

```python
adiv_shannon= alpha_diversity("shannon", arr, simple_ids)
print(adiv_shannon)
```

```python
import pandas as pd
sample_md11 = pd.DataFrame([
    ['HC-1-pre','HC-pre-culture'],
    ['HC-2-pre','HC-pre-culture'],
    ['HC-3-pre','HC-pre-culture'],
    ['OSCC-1-pre','OSCC-pre-culture'],
    ['OSCC-2-pre','OSCC-pre-culture'], 
    ['OSCC-3-pre','OSCC-pre-culture'], 
    ['OSCC-4-pre','OSCC-pre-culture']],
    index=[ 'HC-1-pre', 'HC-2-pre','HC-3-pre','OSCC-1-pre','OSCC-2-pre','OSCC-3-pre','OSCC-4-pre'],
    columns=['SimpleId','Group'])
sample_md11['Adiv_shannon'] = adiv_shannon
sample_md11
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator
palette = ["lightskyblue", "mistyrose"]
tips = sns.load_dataset("tips")
order=['HC-pre-culture','OSCC-pre-culture']
# 绘制箱线图和点图
ax = sns.boxplot(x='Group', y='Adiv_shannon', data=sample_md11, order=order, palette=palette)

xlabel = plt.xlabel('Group', fontsize=15)
plt.xticks(fontsize=9)
sns.stripplot(x='Group', y='Adiv_shannon', data=sample_md11, size=2, jitter=True, edgecolor="black", linewidth=2, order=order)

# 添加注释
annot = Annotator(ax, [('HC-pre-culture','OSCC-pre-culture')],
                  data=sample_md11, x='Group', y='Adiv_shannon', order=order)
annot.configure(loc='outside')
annot.set_custom_annotations(["p=0.503578"])
annot.configure(test='t-test_ind', text_format='full',fontsize='large', loc='inside',
                comparisons_correction=None, line_height=0.05, text_offset=8)
annot.annotate()
ylabel = plt.ylabel('Shannon Diversity', fontsize=15)
# 调整横坐标字体大小
plt.xticks(fontsize=12)
plt.yticks(fontsize=15)
#添加标题HC vs OSCC
plt.title('Shannon Diversity of HC-pre-culture vs OSCC-pre-culture', fontsize=15)
#y轴坐标范围
plt.ylim(2, 4.8)
# 保存图像
plt.savefig('Shannonpre.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
