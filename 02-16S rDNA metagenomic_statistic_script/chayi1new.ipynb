```python
df2=df1[['groups', 'HC-1-pre', 'HC-2-pre','HC-3-pre','HC-1-post','HC-2-post',  'HC-3-post',
         'OSCC-1-pre',  'OSCC-2-pre','OSCC-3-pre','OSCC-4-pre',
       'OSCC-1-post','OSCC-2-post','OSCC-3-post',  'OSCC-4-post']]
df2
```

```python
#df2以groups列为索引
df3 = df2.set_index('groups')
df3
```

```python
# 构建DeseqDataSet 对象
import os
import pickle as pkl
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
from pydeseq2.utils import load_example_data
```

```python
OUTPUT_PATH = "D:\jupy.ipynb\OSCC"
os.makedirs(OUTPUT_PATH, exist_ok=True)
```

```python
samples_to_keep = ~metadata.Group.isna()
counts_df = counts_df.loc[samples_to_keep]
metadata = metadata.loc[samples_to_keep]
```

```python
genes_to_keep = counts_df.columns[counts_df.sum(axis=0) >= 1]
counts_df = counts_df[genes_to_keep]
```

```python
dds = DeseqDataSet(counts=counts_df,metadata=metadata,design_factors=["Group"],refit_cooks=True,n_cpus=8)
dds
```

```python
dds.deseq2()
```

```python
print(dds.varm["LFC"])
```

```python
stat_res_oo_vs_oh = DeseqStats(dds, contrast=["Group", "Pre-Healthy","Post-Healthy"], n_cpus=8)
```

```python
stat_res_oo_vs_oh.summary()
```

```python
stat_res_oo_vs_oh.results_df.to_csv(os.path.join(OUTPUT_PATH, "stat_res_eh_vs_oh1.csv"))
```
